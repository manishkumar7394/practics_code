### firstly we can import all visulization libraries  and load dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
tips = sns.load_dataset('tips')
tips
```

# scatter plot axis label functions


```python
sns.scatterplot(data=tips, x='total_bill', y='tip',hue='sex',style='time',size='size')
```

# relplot for > figure lavel > square shape


```python
sns.relplot(data=tips, x='total_bill', y='tip', kind='scatter',hue='sex',style='time',size='size')
```

```python
# line plot
gap = px.data.gapminder()
temp_df = gap[gap['country'] == 'India']
temp_df
```

```python
#lineplot lavle function
sns.lineplot(data=temp_df, x='year', y = 'lifeExp')
```

```python
sns.relplot(data=temp_df, x='year', y= 'lifeExp',kind= 'lie
```

```python
# hue -> style
temp_df = gap[gap['country'].isin(['India','Brazil','Germany'])]
temp_df
```

```python
sns.relplot(kind = 'line', data=temp_df , x= 'year', y= 'lifeExp', hue= 'country',style ='continent' )
```

```python
# hue -> style
temp_df = gap[gap['country'].isin(['India','China','Pakistan'])]
temp_df
```

```python
sns.relplot(kind= 'line', data=temp_df, x= 'year', y= 'lifeExp',hue= 'country', style= 'iso_num'	)
```
